### Importing libarary 

```python
import pandas as pd
import os
```

### Importing data 

```python
df = pd.read_csv('./SalesAnalysis/Sales_Data/Sales_April_2019.csv')
df.head()

# Head of the sales data in april 
```

### Importing all month data 

```python
files = [file for file in os.listdir('./SalesAnalysis/Sales_Data')]
for file in files:
    print(file)
# Showing all the file in that directory 
```

```python
all_df = pd.DataFrame()
for file in files:
    df = pd.read_csv('./SalesAnalysis/Sales_Data/'+file)
    all_df = pd.concat([all_df, df])

    
# All month data are mereged into one data set 
```

```python
all_df.head()
```

### Checking and droping  NaN values

```python
nan_df = all_df[all_df.isna().any(axis = 1)]
nan_df


# Cleaning data set is being done before anlysis of the data set 
```

```python
all_df.dropna(inplace = True)
```

```python
nan_df = all_df[all_df.isna().any(axis = 1)]
nan_df


# Now all the NaN value are droped
```

```python
all_df = all_df.reset_index(drop = True)

# Data set with no missing values 
```

```python
all_df.head()
```

### Creating a new columns with month only 

```python
Month = all_df['Order Date'].str[0:2]
all_df['Month'] = Month
all_df

# Created a new column of month of purchase 
```

### Some month columns while creating using string method getting 'Or' 

```python
tem = all_df[all_df['Order Date'].str[0:2] == 'Or']
tem

# While creating month column  got some OR so thei need to be eliminated before analysis 
```

### Droping 'Or' from  month columns 

```python
all_df = all_df.loc[all_df['Order Date'].str[0:2] != 'Or']
```

### Changing data type to correct format 

```python
all_df['Month'] = all_df['Month'].astype('int32')
```

```python
all_df.head()
```

```python
all_df.groupby('Month').sum()['Quantity Ordered'].sort_values(ascending = False)

# Hence the column is str type so we need to change to correct formate 
```

```python
all_df['Quantity Ordered'] = pd.to_numeric(all_df['Quantity Ordered'])
all_df['Price Each'] = pd.to_numeric(all_df['Price Each'])


# Changed correct fromat of data type
```

```python
all_df['Sales'] = all_df['Quantity Ordered'] * all_df['Price Each']
```

```python
all_df.head()
```

### Which month has highest sales ?

```python
result = all_df.groupby('Month').sum()
result
```

### What city has the highest number of sales ?

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]


# Created UDF to extract city and state from Purchase address
```

```python
all_df['City'] = all_df['Purchase Address'].apply(lambda x: get_city(x) + ' ( ' + get_state(x) + ' )')
```

```python
all_df.head()
```

```python
City_names = [city for city,df in all_df.groupby('City')]
City_names

# Extracted name of city with state 
```

```python
city_result = all_df.groupby('City').sum()
```

```python
city_result
```

```python

plt.bar(City_names, city_result['Sales'])
plt.xticks(City_names, rotation = 'vertical', size = 8)
plt.xlabel('Cities')
plt.ylabel('Sales in USD')
plt.show()
```

#### San Francisco ( CA)  has a maximum sales

### What products are most often sold together

```python
df = all_df[all_df['Order ID'].duplicated(keep = False)]
df.head(20)
```

```python
df['Grouped'] =  df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df.head(20)
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(20)
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key, ':', value)
```
